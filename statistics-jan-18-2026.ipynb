```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import random

sales = np.array([165,182,140,153,172,168,187,125,212],dtype=int)
hours = np.array([7,8,6,3,2,4,9,12,13])

df=pd.DataFrame({"Sales":sales, "Hours":hours})
print(df)
```

```
sample = [int(x) for x in random.sample(list(sales), 6)]
print(sample)
```

```
mean_sample = np.mean(sample)
mean_sales = np.mean(sales)

median_sample = np.median(sample)
median_sales = np.median(sales)

mode_res1 = stats.mode(sample, keepdims=True)
mode_sample = int(mode_res1.mode[0])
mode_res2 = stats.mode(sales, keepdims=True)
mode_sales = int(mode_res2.mode[0])

print(mean_sample, mean_sales, sep=', ')
print(median_sample, median_sales)
print(mode_sample, mode_sales)
```

```
# Dispersions/Spread
variance_sales=np.var(sales,ddof=1)
std_sales=np.std(sales,ddof=1)
iqr_sales = np.percentile(sales,75)-np.percentile(sales,25)

print(variance_sales, std_sales, iqr_sales, sep=', ')
```

```
# correlation

corr = np.corrcoef(sales,hours)[0,1]
print(corr)

skew_sales=stats.skew(sales)
print(skew_sales)

kurt_sales=stats.kurtosis(sales)
print(kurt_sales)
```

```
plt.scatter(hours, sales)
plt.xlabel("working hours")
plt.ylabel("Sales")
plt.title("Hours Vs Sales")
plt.show()
```
