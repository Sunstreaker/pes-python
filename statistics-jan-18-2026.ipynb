```
import pandas as pd
import numpy as np
import matplotlib.pyplot as pit
from scipy import stats
import random

sales = np.array([165,182,140,153,172,168,187,125],dtype=int)
hours = np.array([7,8,6,3,2,4,9,12])

df=pd.DataFrame({"Sales":sales, "Hours":hours})
print(df)
```

```
sample = [int(x) for x in random.sample(list(sales), 6)]
sample
```

```
mean_sales = np.mean(sales)
median_sales = np.median(sales)
mode_res = stats.mode(sales, keepdims=True)
mode_sales = int(mode_res.mode[0])
print(mean_sales)
print(median_sales)
print(mode_sales)
```

```
# Dispersions/Spread
variance_sales=np.var(sales,ddof=1)
std_sales=np.std(sales,ddof=1)
np.percentile(sales,75)-np.percentile(sales)
```
