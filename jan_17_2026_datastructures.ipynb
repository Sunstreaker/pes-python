```python
p1 = 'c'
p2 = 'c'
print(id(p1))
print(id(p2))
p2 = 'd'
print(id(p2))
p2 = 'c'
print(id(p2))
```

## list operations
- append

```python
l1 = [10, 20, 30, 40, 50]
l1.append(70)
print(l1)
l1.insert(5,60)
print(l1)
l1.extend([80,90])
l2 = [ 'a', 'b', 'c']
print(l1 + l2) # concat
```

## list operations
- sort asc
- sort desc

```python
l1 = [2, 4, 1, 9, 7]
print(l1)
print(l1.sort())
print(l1)
print(l1.sort(reverse=True))
print(l1)
```

```python
l1 = [2, 4, 1, 9, 7]
print(sorted(l1))
print(sorted(l1, reverse=True))
print(l1)
```

## list operations - membership operators
- in
- not in

```python
l1 = [1,2,3,4,5]
print(3 in l1)
print(22 in l1)
print(33 not in l1)
```

```python
l1 = [10, 10, 22, 34, 55]
l2 = [5, 10, 22, 77, 88]
a = []
for l in l1:
    if l in l2:
        a.append(l)
        l2.remove(l)
print(a)
```

## Program
- find the total value

```python
p = ["kholi", 100, 20, 0, 60]
p1 = p[1:]
print(p1)
print(sum(p1))
```

```python
p = ["kholi", 100, 20, 0, 60]
total = 0
for i in p[1:]:
  total += i
print(total)
```
